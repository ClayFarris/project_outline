```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# importing datasets
min = pd.read_csv('MINIMUM_WAGES.csv')
hap15 = pd.read_csv('Happiness2015.csv')
hap16 = pd.read_csv('Happiness2016.csv')
hap17 = pd.read_csv('Happiness2017.csv')
hap18 = pd.read_csv('Happiness2018.csv')
```

```python
# creating DataFrames by selecting countries from 'MINIMUM_WAGES' dataset
# that are also present in the 'Happiness2015-2018" datasets in order to measure variables of interest

min_sub18 = min.iloc[[17,2,18,0,25,27,9,1,15,26,10,4,6,16,3,28,23,30,21,19,14,22,13,11,12,29,5, 8, 24, 20, 7]]

hap2015 = hap15.loc[[4,6,8,9,10,11,13,14,15,16,17,18,20,26,28,30,32,35,44,45,46,54,55,59,63,72,75,87,88,101,103]]

hap2016 = hap16.loc[[5,6,7,8,10,12,13,16,17,18,19,20,22,23,26,30,31,36,44,52,55,56,57,59,62,67,71,77,90,93,98]]

hap2017 = hap17.loc[[5,6,7,9,10,11,13,14,16,17,18,19,21,22,24,30,33,35,39,45,48,50,51,53,54,61,65,68,74,86,88]]

hap2018 = hap18.iloc[[5,6,7,9,10,12,13,15,16,17,18,20,22,23,24,27,35,36,38,41,49,50,52,53,56,58,62,68,73,76,78]]




```

```python
min_sub18.head()
```

```python
# creating new dataframe by slicing columns containing the relevant years
min = min_sub18.loc[:, ["Country","2015", "2016", "2017", "2018"]]

min.head()
```

```python
# creating new dataframe by slicing columns containing 
# the relevant years and calculating the average wages 
# for each Country
min = min_sub18.loc[:, ["Country","2015", "2016", "2017", "2018"]]

min["Mean"] = min.mean(axis=1)

min.head()
```

```python
# Making the country column the index
min = min.set_index("Country")
```

```python
min
```

```python
# Creating a Series with countries and their average minimum wages
min_avg = min.loc[:,["Mean"]]

min_avg
```

```python
# Bringing in happiness data
hap2015.head()
```

```python
# Creating Series from Happiness data from 2015 using the "Country" and "Happiness Score Columns"
happy15 = hap2015.loc[:,["Country", "Happiness Score"]]
happy15 = happy15.set_index("Country")


happy15.head()
```

```python
hap2016.head()
```

```python
happy16 = hap2016.loc[:,["Country", "Happiness Score"]]
happy16 = happy16.set_index("Country")
                            
happy16.head()                           
```

```python
hap2017.head()
```

```python
happy17 = hap2017.loc[:,["Country", "Happiness.Score"]]
happy17 = happy17.set_index("Country")

happy17.head()
```

```python
hap2018.head()
```

```python
happy18 = hap2018.loc[:,["Country or region", "Score"]]
happy18 = happy18.set_index("Country or region")

happy18.head()
```

```python
happy17 = happy17.rename(columns={"Happines.Score":"Happiness Score"})
```

```python
happy18_new = happy18.rename(columns={"Score" : "Happiness Score"}, index = {"Country or region" : "Country"})
```

```python
happy18_new.head()
```

```python
happiness= pd.concat([happy15, happy16, happy17, happy18], axis=1)
```

```python
happiness
```

```python
# Finding average happiness score for each country for the years 2015-2018
happiness["Mean"] = happiness.mean(axis=1)
happiness
```

```python
# creating a series from the average happines scores 
happiness_avg = happiness.loc[:,["Mean"]]
```

```python
happiness_avg
```

```python
# Creating a dataframe using average happiness scores and average minimum wages from 2015-2018
minhap = pd.concat([happiness_avg, min_avg], axis=1)
```

```python
minhap
```

```python
# Changing country names that are same country but under 
# different names in respective data sets to avoid generating 
# null values like in the above dataframe.
min_avg
```

```python
min_avg = min_avg.rename(index={"Slovak Republic":"Slovakia", "Korea":"South Korea", "Russian Federation":"Russia"})
```

```python
min_avg
```

```python
# Creating updated dataframe that has removed null values and replaced them with the proper data
minhap = pd.concat([min_avg, happiness_avg], axis=1)
```

```python
#renaming variables
```

```python
minhap=minhap.rename_axis('Country').set_axis(['Wages', 'Happiness Score'], axis=1)
```

```python
minhap.plot(kind='bar', x='Wages', y='Happiness Score', color='green')
plt.title('Wages and Happiness')
plt.xlabel('Average Minimum Wage 2015-2018')
plt.ylabel('Average Happiness Score 2015-2018')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.bar(minhap['Country'], minhap['Wages'], color='tab:blue', alpha=0.7, label='Wages')
ax.set_ylabel('Wages')
ax.set_xticklabels(minhap['Country'], rotation=90, fontsize=12)

ax2 = ax.twinx()
ax2.bar(minhap['Country'], minhap['Happiness Score'], color='tab:orange', alpha=0.7, label='Happiness Score', width=0.5)
ax2.set_ylabel('Happiness Score')
ax2.set_ylim(0, 8)

ax.set_xlim(-0.5, len(minhap['Country'])-0.5)

plt.title('Wages and Happiness Score by Country', fontsize=16)
plt.legend(loc='upper left')
plt.show()
```
